```python
# Variavel Independente
X = np.array(data['Ida_IMT'])

# Variavel Dependente
y = np.array(data['Volta_IMT'])

# Correlacao
correlacao = np.corrcoef(X, y)

print(f'X: {X}\n')
print(f'y: {y}\n')
print(f'Correlacao: {correlacao}')
```

```python
X = X.reshape(-1, 1)
X
```

```python
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
plt.scatter(X, y, marker = '.', color='black')
plt.title('Tempo de ida ao IMT x Tempo de volta do IMT')
plt.xlabel('Tempo de ida ao IMT (min)')
plt.ylabel('Tempo de Volta do IMT (min)')

plt.plot(X, modelo.predict(X), color='red')
```

```python
modelo.predict([[100]])
```

# **Visualizando residuais**

```python
visualizar_residuais = ResidualsPlot(modelo)
visualizar_residuais.fit(X, y)

visualizar_residuais.poof()
```
